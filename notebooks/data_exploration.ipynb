## Build a Reusable Pitcher Projection Function

So far, the notebook only worked for one pitcher because the data was stored in variables like `skubal`.

This next section creates a reusable function that can:
- look up a pitcher by name
- pull that pitcher's Statcast data
- remove spring training games
- calculate strikeouts by game
- calculate recent strikeout averages
- create a simple projected strikeout number

```python
from pybaseball import playerid_lookup, statcast_pitcher

def project_pitcher_k(last_name, first_name, season):
    player_ids = playerid_lookup(last_name, first_name)
    mlbam_id = int(player_ids.iloc[0]['key_mlbam'])

    start_date = f'{season}-03-27'
    end_date = f'{season}-10-01'

    data = statcast_pitcher(start_date, end_date, mlbam_id)

    games = (
        data.groupby('game_date')
        .agg(
            strikeouts=('events', lambda x: (x == 'strikeout').sum()),
            pitches=('pitch_type', 'count'),
            avg_velocity=('release_speed', 'mean')
        )
        .reset_index()
        .sort_values('game_date')
    )

    games['last_3_avg_k'] = games['strikeouts'].rolling(3).mean()
    games['last_5_avg_k'] = games['strikeouts'].rolling(5).mean()
    games['season_avg_k'] = games['strikeouts'].expanding().mean()

    latest = games.iloc[-1]

    projected_k = (
        latest['last_3_avg_k'] * 0.4 +
        latest['last_5_avg_k'] * 0.4 +
        latest['season_avg_k'] * 0.2
    )

    return {
        'pitcher': f'{first_name.title()} {last_name.title()}',
        'mlbam_id': mlbam_id,
        'season': season,
        'projected_k': projected_k,
        'games': games
    }
```

That would lean toward the over.

If our projection is lower than the sportsbook line, that would lean toward the under.

## Create a Simple Betting Lean

This step turns the projection difference into a simple recommendation.

For now, the logic is:

- If our projection is at least 0.5 strikeouts above the line, lean Over
- If our projection is at least 0.5 strikeouts below the line, lean Under
- If the difference is smaller than 0.5, call it No Bet

This keeps us from forcing a bet when the projection and sportsbook line are too close.

```python
if difference >= 0.5:
    recommendation = 'Over'
elif difference <= -0.5:
    recommendation = 'Under'
else:
    recommendation = 'No Bet'

recommendation
```

## Create a Reusable Sportsbook Comparison Function

Now we turn the sportsbook comparison into a reusable function.

This function takes:

- a pitcher projection result
- a sportsbook strikeout line

It returns a clean summary table with:

- pitcher name
- projected strikeouts
- sportsbook line
- difference
- recommendation

All displayed numbers are rounded to hundredths.

```python
def compare_to_line(result, sportsbook_line):
    projected_k = round(result['projected_k'], 2)
    difference = round(projected_k - sportsbook_line, 2)

    if difference >= 0.5:
        recommendation = 'Over'
    elif difference <= -0.5:
        recommendation = 'Under'
    else:
        recommendation = 'No Bet'

    summary = pd.DataFrame([{
        'pitcher': result['pitcher'],
        'projected_k': projected_k,
        'sportsbook_line': sportsbook_line,
        'difference': difference,
        'recommendation': recommendation
    }])

    return summary
```

## Add Edge Strength

The difference between our projection and the sportsbook line is called the edge.

A bigger edge means our projection disagrees more strongly with the sportsbook line.

For this first version, we will label each edge like this:

- Strong Edge: difference is 1.5 strikeouts or more
- Medium Edge: difference is 1.0 to 1.49 strikeouts
- Small Edge: difference is 0.5 to 0.99 strikeouts
- No Bet: difference is less than 0.5 strikeouts

```python
def label_edge(difference):
    abs_difference = abs(difference)

    if abs_difference >= 1.5:
        return 'Strong Edge'
    elif abs_difference >= 1.0:
        return 'Medium Edge'
    elif abs_difference >= 0.5:
        return 'Small Edge'
    else:
        return 'No Bet'
```

## Add Sample Size Awareness

A projection is more trustworthy when it is based on enough starts.

For this first version, we will count how many games were used for each pitcher projection.

Then we will label the sample size:

- Good Sample: 10 or more starts
- Small Sample: fewer than 10 starts

This helps us avoid trusting projections that are based on too little data.

```python
def project_pitcher_k(last_name, first_name, season):
    player_ids = playerid_lookup(last_name, first_name)
    mlbam_id = int(player_ids.iloc[0]['key_mlbam'])

    start_date = f'{season}-03-27'
    end_date = f'{season}-10-01'

    data = statcast_pitcher(start_date, end_date, mlbam_id)

    games = (
        data.groupby('game_date')
        .agg(
            strikeouts=('events', lambda x: (x == 'strikeout').sum()),
            pitches=('pitch_type', 'count'),
            avg_velocity=('release_speed', 'mean')
        )
        .reset_index()
        .sort_values('game_date')
    )

    games['last_3_avg_k'] = games['strikeouts'].rolling(3).mean()
    games['last_5_avg_k'] = games['strikeouts'].rolling(5).mean()
    games['season_avg_k'] = games['strikeouts'].expanding().mean()

    latest = games.iloc[-1]

    projected_k = (
        latest['last_3_avg_k'] * 0.4 +
        latest['last_5_avg_k'] * 0.4 +
        latest['season_avg_k'] * 0.2
    )

    return {
        'pitcher': f'{first_name.title()} {last_name.title()}',
        'mlbam_id': mlbam_id,
        'season': season,
        'projected_k': projected_k,
        'games_used': len(games),
        'games': games
    }
```

## Update the Sportsbook Comparison Function

Now we update the sportsbook comparison function so the output includes sample size.

The function will still compare our projected strikeouts to the sportsbook line, but now it will also show how many games were used and whether that is a good or small sample.

```python
def compare_to_line(result, sportsbook_line):
    projected_k = round(result['projected_k'], 2)
    difference = round(projected_k - sportsbook_line, 2)

    if difference >= 0.5:
        recommendation = 'Over'
    elif difference <= -0.5:
        recommendation = 'Under'
    else:
        recommendation = 'No Bet'

    if result['games_used'] >= 10:
        sample_size = 'Good Sample'
    else:
        sample_size = 'Small Sample'

    summary = pd.DataFrame([{
        'pitcher': result['pitcher'],
        'games_used': result['games_used'],
        'projected_k': projected_k,
        'sportsbook_line': sportsbook_line,
        'difference': difference,
        'recommendation': recommendation,
        'sample_size': sample_size
    }])

    return summary
```

## Add Final Bet or Pass Logic

Now we create a final action column.

The recommendation column says whether the projection leans Over, Under, or No Bet.

The final action column is stricter. It only says to bet when:

- the sample size is good
- the edge is at least medium
- the recommendation is Over or Under

Otherwise, the action is Pass.

```python
def final_action(row):
    if (
        row['sample_size'] == 'Good Sample'
        and row['edge_strength'] in ['Medium Edge', 'Strong Edge']
        and row['recommendation'] == 'Over'
    ):
        return 'Bet Over'

    elif (
        row['sample_size'] == 'Good Sample'
        and row['edge_strength'] in ['Medium Edge', 'Strong Edge']
        and row['recommendation'] == 'Under'
    ):
        return 'Bet Under'

    else:
        return 'Pass'
```

## Apply Final Bet or Pass Logic

Now we apply the final action function to every row in the pitcher board.

This gives us a clearer final decision for each pitcher.

```python
pitcher_board['final_action'] = pitcher_board.apply(final_action, axis=1)

pitcher_board
```

## Create a Full Pitcher Board Function

Now we combine the projection, sportsbook comparison, edge strength, and final action logic into one reusable function.

This function takes a list of pitchers and lines, then returns a complete pitcher board.

Each pitcher needs:

- first name
- last name
- sportsbook strikeout line

The function will:

1. Project each pitcher's strikeouts
2. Compare the projection to the sportsbook line
3. Add edge strength
4. Add final bet/pass action
5. Return one clean board

```python
def build_pitcher_board(pitchers, season):
    all_summaries = []

    for pitcher in pitchers:
        try:
            result = project_pitcher_k(
                pitcher['last_name'],
                pitcher['first_name'],
                season
            )

            summary = compare_to_line(result, pitcher['line'])
            all_summaries.append(summary)

        except Exception as e:
            print(f"Error with {pitcher['first_name']} {pitcher['last_name']}: {e}")

    board = pd.concat(all_summaries, ignore_index=True)

    board['edge_strength'] = board['difference'].apply(label_edge)
    board['final_action'] = board.apply(final_action, axis=1)

    board['abs_difference'] = board['difference'].abs()
    board = board.sort_values('abs_difference', ascending=False)
    board = board.drop(columns=['abs_difference'])

    return board
```

## Handle Pitchers With Missing Data

Some pitchers may not have data for the selected season.

This can happen if the pitcher was injured, did not pitch, changed roles, or if the name lookup does not return usable data.

Instead of skipping those pitchers, we will add them to the board with a clear status of No Data.

```python
def build_pitcher_board(pitchers, season):
    all_summaries = []

    for pitcher in pitchers:
        try:
            result = project_pitcher_k(
                pitcher['last_name'],
                pitcher['first_name'],
                season
            )

            summary = compare_to_line(result, pitcher['line'])
            all_summaries.append(summary)

        except Exception as e:
            missing_summary = pd.DataFrame([{
                'pitcher': f"{pitcher['first_name'].title()} {pitcher['last_name'].title()}",
                'games_used': 0,
                'projected_k': None,
                'sportsbook_line': pitcher['line'],
                'difference': None,
                'recommendation': 'No Data',
                'sample_size': 'No Data',
                'error_message': str(e)
            }])

            all_summaries.append(missing_summary)

    board = pd.concat(all_summaries, ignore_index=True)

    board['edge_strength'] = board['difference'].apply(
        lambda x: 'No Data' if pd.isna(x) else label_edge(x)
    )

    board['final_action'] = board.apply(
        lambda row: 'Pass' if row['recommendation'] == 'No Data' else final_action(row),
        axis=1
    )

    board['abs_difference'] = board['difference'].abs()
    board = board.sort_values('abs_difference', ascending=False, na_position='last')
    board = board.drop(columns=['abs_difference'])

    return board
```

## Fix Sorting When Some Pitchers Have No Data

When a pitcher has no data, the difference value is missing.

To keep the board sortable, missing numeric values should be stored as `np.nan` instead of `None`.

This lets pandas treat the column as numeric and sort missing values last.

```python
def build_pitcher_board(pitchers, season):
    all_summaries = []

    for pitcher in pitchers:
        try:
            result = project_pitcher_k(
                pitcher['last_name'],
                pitcher['first_name'],
                season
            )

            summary = compare_to_line(result, pitcher['line'])
            all_summaries.append(summary)

        except Exception as e:
            missing_summary = pd.DataFrame([{
                'pitcher': f"{pitcher['first_name'].title()} {pitcher['last_name'].title()}",
                'games_used': 0,
                'projected_k': np.nan,
                'sportsbook_line': pitcher['line'],
                'difference': np.nan,
                'recommendation': 'No Data',
                'sample_size': 'No Data',
                'error_message': str(e)
            }])

            all_summaries.append(missing_summary)

    board = pd.concat(all_summaries, ignore_index=True)

    board['edge_strength'] = board['difference'].apply(
        lambda x: 'No Data' if pd.isna(x) else label_edge(x)
    )

    board['final_action'] = board.apply(
        lambda row: 'Pass' if row['recommendation'] == 'No Data' else final_action(row),
        axis=1
    )

    board['abs_difference'] = board['difference'].abs()
    board = board.sort_values('abs_difference', ascending=False, na_position='last')
    board = board.drop(columns=['abs_difference'])

    return board
```

## Add a Status Column

Now we add a status column to make the board easier to interpret.

The status column summarizes each row:

- Playable: the final action is Bet Over or Bet Under
- Pass - Small Edge: the projection is too close to the sportsbook line
- Pass - Small Sample: there are not enough starts
- No Data: the pitcher does not have usable data

```python
def add_status(row):
    if row['recommendation'] == 'No Data':
        return 'No Data'

    if row['final_action'] in ['Bet Over', 'Bet Under']:
        return 'Playable'

    if row['sample_size'] == 'Small Sample':
        return 'Pass - Small Sample'

    if row['edge_strength'] == 'Small Edge' or row['edge_strength'] == 'No Bet':
        return 'Pass - Small Edge'

    return 'Pass'
```

Now we update the full board function so every pitcher gets a status label.

This makes the final board easier to scan.

## Make the Projection Function Season Flexible

The current projection function uses a hardcoded regular season start date.

That works for 2025 because the regular season started on March 27, 2025.

To make the project work for future seasons, including 2026, we will update the function so the regular season start date can be entered as an input.

```python
def project_pitcher_k(last_name, first_name, season, regular_season_start):
    player_ids = playerid_lookup(last_name, first_name)
    mlbam_id = int(player_ids.iloc[0]['key_mlbam'])

    start_date = regular_season_start
    end_date = f'{season}-10-01'

    data = statcast_pitcher(start_date, end_date, mlbam_id)

    games = (
        data.groupby('game_date')
        .agg(
            strikeouts=('events', lambda x: (x == 'strikeout').sum()),
            pitches=('pitch_type', 'count'),
            avg_velocity=('release_speed', 'mean')
        )
        .reset_index()
        .sort_values('game_date')
    )

    games['last_3_avg_k'] = games['strikeouts'].rolling(3).mean()
    games['last_5_avg_k'] = games['strikeouts'].rolling(5).mean()
    games['season_avg_k'] = games['strikeouts'].expanding().mean()

    latest = games.iloc[-1]

    projected_k = (
        latest['last_3_avg_k'] * 0.4 +
        latest['last_5_avg_k'] * 0.4 +
        latest['season_avg_k'] * 0.2
    )

    return {
        'pitcher': f'{first_name.title()} {last_name.title()}',
        'mlbam_id': mlbam_id,
        'season': season,
        'regular_season_start': regular_season_start,
        'projected_k': projected_k,
        'games_used': len(games),
        'games': games
    }
```

## Update the Board Function for Season Start Date

## Add Projection Ingredients to the Board

The projected strikeout number comes from three averages:

- last 3 starts average strikeouts
- last 5 starts average strikeouts
- season average strikeouts

Adding these columns to the board makes the projection easier to understand and review.

```python
def project_pitcher_k(last_name, first_name, season, regular_season_start):
    player_ids = playerid_lookup(last_name, first_name)
    mlbam_id = int(player_ids.iloc[0]['key_mlbam'])

    start_date = regular_season_start
    end_date = f'{season}-10-01'

    data = statcast_pitcher(start_date, end_date, mlbam_id)

    games = (
        data.groupby('game_date')
        .agg(
            strikeouts=('events', lambda x: (x == 'strikeout').sum()),
            pitches=('pitch_type', 'count'),
            avg_velocity=('release_speed', 'mean')
        )
        .reset_index()
        .sort_values('game_date')
    )

    games['last_3_avg_k'] = games['strikeouts'].rolling(3).mean()
    games['last_5_avg_k'] = games['strikeouts'].rolling(5).mean()
    games['season_avg_k'] = games['strikeouts'].expanding().mean()

    latest = games.iloc[-1]

    projected_k = (
        latest['last_3_avg_k'] * 0.4 +
        latest['last_5_avg_k'] * 0.4 +
        latest['season_avg_k'] * 0.2
    )

    return {
        'pitcher': f'{first_name.title()} {last_name.title()}',
        'mlbam_id': mlbam_id,
        'season': season,
        'regular_season_start': regular_season_start,
        'projected_k': projected_k,
        'last_3_avg_k': latest['last_3_avg_k'],
        'last_5_avg_k': latest['last_5_avg_k'],
        'season_avg_k': latest['season_avg_k'],
        'games_used': len(games),
        'games': games
    }
```

```python
def compare_to_line(result, sportsbook_line):
    projected_k = round(result['projected_k'], 2)
    difference = round(projected_k - sportsbook_line, 2)

    if difference >= 0.5:
        recommendation = 'Over'
    elif difference <= -0.5:
        recommendation = 'Under'
    else:
        recommendation = 'No Bet'

    if result['games_used'] >= 10:
        sample_size = 'Good Sample'
    else:
        sample_size = 'Small Sample'

    summary = pd.DataFrame([{
        'pitcher': result['pitcher'],
        'games_used': result['games_used'],
        'last_3_avg_k': round(result['last_3_avg_k'], 2),
        'last_5_avg_k': round(result['last_5_avg_k'], 2),
        'season_avg_k': round(result['season_avg_k'], 2),
        'projected_k': projected_k,
        'sportsbook_line': sportsbook_line,
        'difference': difference,
        'recommendation': recommendation,
        'sample_size': sample_size
    }])

    return summary
```

```python
def build_pitcher_board(pitchers, season, regular_season_start):
    all_summaries = []

    for pitcher in pitchers:
        try:
            first_name, last_name = split_pitcher_name(pitcher['pitcher'])

            result = project_pitcher_k(
                last_name,
                first_name,
                season,
                regular_season_start
            )

            summary = compare_to_line(result, pitcher['line'])
            summary['opponent'] = pitcher['opponent']
            summary['home_away'] = pitcher['home_away']

            all_summaries.append(summary)

        except Exception as e:
            missing_summary = pd.DataFrame([{
                'pitcher': pitcher['pitcher'],
                'opponent': pitcher['opponent'],
                'home_away': pitcher['home_away'],
                'games_used': 0,
                'last_3_avg_k': np.nan,
                'last_5_avg_k': np.nan,
                'season_avg_k': np.nan,
                'projected_k': np.nan,
                'sportsbook_line': pitcher['line'],
                'difference': np.nan,
                'recommendation': 'No Data',
                'sample_size': 'No Data',
                'error_message': str(e)
            }])

            all_summaries.append(missing_summary)

    board = pd.concat(all_summaries, ignore_index=True)

    board['edge_strength'] = board['difference'].apply(
        lambda x: 'No Data' if pd.isna(x) else label_edge(x)
    )

    board['final_action'] = board.apply(
        lambda row: 'Pass' if row['recommendation'] == 'No Data' else final_action(row),
        axis=1
    )

    board['status'] = board.apply(add_status, axis=1)

    board['abs_difference'] = board['difference'].abs()
    board = board.sort_values('abs_difference', ascending=False, na_position='last')
    board = board.drop(columns=['abs_difference'])

    columns = [
        'pitcher',
        'opponent',
        'home_away',
        'games_used',
        'last_3_avg_k',
        'last_5_avg_k',
        'season_avg_k',
        'projected_k',
        'sportsbook_line',
        'difference',
        'recommendation',
        'sample_size',
        'edge_strength',
        'final_action',
        'status'
    ]

    return board[columns]
```

Strikeout projections depend on opportunity.
